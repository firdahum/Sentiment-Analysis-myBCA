## **Import Library**


```
from google_play_scraper import app, reviews_all, Sort # Mengambil data aplikasi dan ulasan
import pandas as pd
import csv
```

## **Scapping Data**

```
# Ambil semua ulasan dari aplikasi dengan ID tertentu
scrapreview = reviews_all(
    'com.bca.mybca.omni.android',  # ID aplikasi
    lang='id',                     # Bahasa ulasan
    country='id',                  # Negara asal ulasan
    sort=Sort.MOST_RELEVANT,       # Urutan ulasan
    count=11000                    # Maks jumlah ulasan
)
```

```
# Menyimpan ke CSV
import csv

with open('ulasan_sentimen.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review'])  # Menulis header kolom
    for review in scrapreview:
        writer.writerow([review['content']])  # Menulis konten ulasan ke dalam file CSV
```

```
mybca_reviews_df = pd.DataFrame(scrapreview)
mybca_reviews_df.shape
mybca_reviews_df.head()
mybca_reviews_df.to_csv('ulasan_mybca.csv', index=False)
```
